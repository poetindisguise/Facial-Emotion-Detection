# Genetic Algorithm Functions

## Fitness Function

```
#to calculate (accuracy = fitness function) using knn classifier 
def fitness(feature_subset):
  all_columns = np.array(combined_df.columns)
  required_columns = list()

  count = 0
  for i in feature_subset:
    if(i == '1'):
      required_columns.append(all_columns[count])
    count += 1

  df_required = combined_df[required_columns]

  df_required_x = df_required.iloc[:,:-1]
  df_required_y = df_required.iloc[:,-1]
  
  from sklearn.neighbors import KNeighborsClassifier
  classifier_knn = KNeighborsClassifier(n_neighbors = 4)
  classifier_knn.fit(df_required_x,df_required_y)
  accuracy = classifier_knn.score(df_required_x,df_required_y)

  return accuracy
```

## Single Point Crossover

```
#using single point crossover
#takes inputs as strings and returns strings

import random
def single_pt_crossover(feature_subset_1 , feature_subset_2 ): 
  feature_length = len(feature_subset_1)

  split_index = random.randint(1, feature_length-1)

  return '{}{}'.format(feature_subset_1[:split_index], feature_subset_2[split_index:]), '{}{}'.format(feature_subset_2[:split_index],feature_subset_1[split_index:])

# eg feature_subset_1 = "01110101010" , feature_subset_2 = "10110110101"
# if split_index = 4
#values returned are "01110110101" , "10110101010"
```

## Multi Point Crossover

```
#using multi point crossover
#takes inputs as strings and returns strings

import random
def multi_pt_crossover(feature_subset_1 , feature_subset_2 ): 

  feature_length = len(feature_subset_1)
  split_index_1 = random.randint(1, feature_length-1)
  split_index_2 = random.randint(1, feature_length-1)

  if(split_index_1 <= split_index_2):
    split_index_2 = split_index_2 + 1
  else:
    #swap
    split_index_1 , split_index_2 = split_index_2 , split_index_1

  print(split_index_1)
  print(split_index_2)
  return '{}{}{}'.format(feature_subset_1[:split_index_1],feature_subset_2[split_index_1:split_index_2] ,feature_subset_1[split_index_2:]), '{}{}{}'.format(feature_subset_2[:split_index_1],feature_subset_1[split_index_1:split_index_2] ,feature_subset_2[split_index_2:])
  
```

## Mutation 

```
#using bit swap
def mutation(feature_set):
  str =""
  for i in feature_set:
    if (i == '0'): 
      str += "{}".format('1')
    if(i == '1'):
      str += "{}".format('0')
  return str
```

## Initial Population

```
import random

def initial_population(no_of_features ,no_of_individuals = 1000 ):
  
  #list of individual is population
  population = list() 
  
  for i in range(no_of_individuals):
    individual=""
    for i in range(no_of_features):
      individual = individual + "{}".format(random.randint(0,1))
    population.append(individual)

  return population
  
```

# Determining Best Feature Set

```
import time

def generation(initial_generation,crossover_rate = 0.78 , mutation_rate = 0.08 , no_of_generations = 10 , no_of_offsprings_each_generation =10 ):
#crossover_rate = 0.78 means 78% of offsprings in new_genration is due to crossover 
#mutation_rate = 0.08 means 8% of offsprings in new_generation is due to mutation

  for generation in range(1,no_of_generations+1):
    print('Generation : {}'.format(generation))
    print('')

    new_generation = list()
    best_accuracy_in_each_generation = list()

    #crossover
    for i in range(int(crossover_rate*no_of_offsprings_each_generation)):
      random_no_1 = random.randint(0,no_of_offsprings_each_generation)
      random_no_2 = random.randint(0,no_of_offsprings_each_generation)
      offspring_1 , offspring_2 = single_pt_crossover(initial_generation[random_no_1][1],initial_generation[random_no_2][1])
      fitness_offspring_1 = fitness(offspring_1 + '{}'.format(1))
      fitness_offspring_2 = fitness(offspring_2 + '{}'.format(1))
      new_generation.append([fitness_offspring_1,offspring_1])
      new_generation.append([fitness_offspring_2,offspring_2])
    
    for i in range(int((1-crossover_rate)*no_of_offsprings_each_generation)):
      random_no = random.randint(0,no_of_offsprings_each_generation)
      new_generation.append(initial_generation[random_no])

    #mutation
    for i in range(int(mutation_rate*no_of_offsprings_each_generation)):
      random_no = random.randint(0,no_of_offsprings_each_generation)
      offspring = mutation(new_generation[random_no][1])
      fitness_offspring = fitness(offspring + '{}'.format(1))
      new_generation[random_no] = [fitness_offspring,offspring]

    #sorting
    new_generation = sorted(new_generation,key=lambda x : x[0],reverse=True) 
    for i in new_generation:
      print(i)

    #printing best accuracy and respective subset
    print('')
    print('best accuracy in this generation: {}'.format(new_generation[0][0]))
    best_accuracy_in_each_generation.append(new_generation[0][0])
    print('respective feature subset: {}'.format(new_generation[0][1]))
    print('')

    #if we get accuaracy >0.99 before no_of_generation ends then break
    if(new_generation[0][0] > 0.99):
      print('break because we got a feature subset which gives accuracy > 0.99')
      break

    initial_generation = new_generation

  return new_generation[0][1]
  
```

```
initial_feature_set = initial_population(710,1000)
accuracy_list = list()

for i in initial_feature_set:
  feature_set = i + '{}'.format(1)   #adding 1 at the end so that target variable is always included
  accuracy_list.append(fitness(feature_set))

#converting lists into numpyp array because they are fast
initial_feature_set = np.array(initial_feature_set)
accuracy_list = np.array(accuracy_list)

#appending accuracy and respective feature set
accuracy_feature_set = list()
for i in range(len(accuracy_list)):
  pair = list()
  pair.append(accuracy_list[i])
  pair.append(initial_feature_set[i])
  accuracy_feature_set.append(pair)

#sorting accuracy_feature_set in descending order i.e. max accuracy feature set at the first
sorted_accuracy_feature_set = sorted(accuracy_feature_set,key=lambda x : x[0],reverse=True) 
print('Generation : 0 (Initial Random Generation)')
print("")
for i in sorted_accuracy_feature_set:
  print(i)
print('')

start = time.time()
best_feature_set = generation(sorted_accuracy_feature_set,no_of_offsprings_each_generation=40,no_of_generations=60,crossover_rate=0.8,mutation_rate=0.1)
end = time.time()
print("Time Taken : {}".format(end-start))
```
